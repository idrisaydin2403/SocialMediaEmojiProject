```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\İdris Aydın\\Desktop\\Student_performance_data _.csv")
df
```

```python
df['Gender'] = df['Gender'].replace({0:'Male', 1:'Female'})
df['Ethnicity'] = df['Ethnicity'].replace({0:'Caucasian', 1:'African American', 2:'Asian', 3:'Other'})
df['ParentalEducation'] = df['ParentalEducation'].replace({0:None, 1:'High School', 2:'Some College', 3:'Bachelors', 4:'Higher'})
df['GradeClass'] = df['GradeClass'].replace({0:'A', 1:'B', 2:'C', 3:'D', 4:'F'})
df['Tutoring'] = df['Tutoring'].replace({1:'YES', 0:'NO'})
df
```

```python
print("\nTOTAL NUMBER OF STUDENTS:", len(df))
print("NUMBER OF MALE STUDENTS", len(df[df['Gender'] == 'Male']))
print("NUMBER OF MALE STUDENTS", len(df[df['Gender'] == 'Female']), "\n")

genders = df['Gender'].unique()
numberOfGenders = [(len(df[df['Gender'] == 'Male'])), (len(df[df['Gender'] == 'Female']))]

plt.pie(numberOfGenders, labels=genders, autopct='%1.1f%%', startangle= 90, colors = ['hotpink','blue'])
plt.title("STUDENTS RATIO")
plt.show()
```

```python
ages = [15, 16, 17, 18]
man, fan = [], []
for i in range(15, 19):
    man.append(len(df[(df['Gender'] == 'Male') & (df['Age'] == i)]))
    fan.append(len(df[(df['Gender'] == 'Female') & (df['Age'] == i)]))
    
bar_width = 0.23
    
r1 = np.arange(len(ages))
r2 = [x + bar_width for x in r1]

plt.bar(r1, man, color = 'blue', width = 0.23, label = 'Male')
plt.bar(r2, fan, color = 'hotpink', width= 0.23, label = 'Female') 

plt.title("STUDENT NUMBERS BY AGE CATEGORIES\n")
plt.xlabel("\nAGE")
plt.ylabel("STUDENT NUMBERS\n")
plt.xticks([(r + bar_width / 2) for r in range(len(ages))], ages)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
studentNumbers = []
for e in df['Ethnicity'].unique():
    studentNumbers.append(len(df[df['Ethnicity'] == e]))

plt.bar(df['Ethnicity'].unique(), studentNumbers, width = 0.5, color = 'gray')
plt.title("STUDENT NUMBERS OF BY ETHNIC ORIGIN\n")
plt.xlabel("\nETHNIC")
plt.ylabel("STUDENT NUMBER\n")

plt.show()
```

```python
female_grades = df[(df['Gender'] == 'Female') & (df['GradeClass'].isin(['A','B','C','D','F']))]
resultFemaleGrades = female_grades.groupby(['GradeClass']).size().reset_index(name='Count')

print("\n     FEMALE'S NOTES")
resultFemaleGrades
```

```python
male_grades = df[(df['Gender'] == 'Male') & (df['GradeClass'].isin(['A', 'B', 'C', 'D', 'F']))]
resultMaleGrades = male_grades.groupby(['GradeClass']).size().reset_index(name = 'Count')
 
print("\n      MALE'S NOTES")
resultMaleGrades
```

```python
df.groupby('Ethnicity')['GPA'].mean().reset_index()

ethnicities = df['Ethnicity'].unique()
ethnicities_means = df.groupby('Ethnicity')['GPA'].mean().values

plt.bar(ethnicities, ethnicities_means, color = 'brown', width = 0.7)
plt.title("GRAPH OF GPA OF ETHNIC ORIGIN\n")
plt.xlabel("\nETHNIC")
plt.ylabel("GPA\n")

plt.show()
```

```python
male_stw_mean = round(df[df['Gender'] == 'Male']['StudyTimeWeekly'].mean(), 2)
female_stw_mean = round(df[df['Gender'] == 'Female']['StudyTimeWeekly'].mean(), 2)

print("AVERAGE TOTAL WORKING HOURS MALE'S WEEKLY:",male_stw_mean)
print("AVERAGE TOTAL WORKING HOURS FEMALE'S WEEKLY",female_stw_mean)

plt.bar(['MALE','FEMALE'], [male_stw_mean, female_stw_mean], color = ['blue', 'pink'], width = 0.5)
plt.title("WEEKLY AVERAGE WORK GRAPH\n")
plt.xlabel("\nGENDER")
plt.ylabel("HOUR\n")

plt.show()
```

```python
maleTutoringNumber = len(df[(df['Gender'] == 'Male') & (df['Tutoring'] == 'YES')])
femaleTutoringNumber = len(df[(df['Gender'] == 'Female') & (df['Tutoring'] == 'YES')])

ratio_maleTutoring = round((maleTutoringNumber / len(df[df['Gender'] == 'Male'])) * 100, 2)
ratio_femaleTutoring = round((femaleTutoringNumber / len(df[df['Gender'] == 'Female'])) * 100, 2)

plt.pie([ratio_maleTutoring, ratio_femaleTutoring], labels = ['Male','Female'], startangle = 90, autopct='%1.1f%%')
plt.title("\n\nRATES OF STUDENTS ATTENTION TO PRIVATE LESSONS")
plt.show()
```

```python
maleAverageSuccess = round(df[df['Gender'] == 'Male']['GPA'].mean(), 2)
femaleAverageSuccess = round(df[df['Gender'] == 'Female']['GPA'].mean(), 2)

print('\nAverage success of male students:', maleAverageSuccess)
print('Average success of female students:', femaleAverageSuccess,"\n")

m1 = round(df[(df['Gender'] == 'Male') & (df['Tutoring'] == 'YES')]['GPA'].mean(), 2)
m2 = round(df[(df['Gender'] == 'Male') & (df['Tutoring'] == 'NO')]['GPA'].mean(), 2)

f1 = round(df[(df['Gender'] == 'Female') & (df['Tutoring'] == 'YES')]['GPA'].mean(), 2)
f2 = round(df[(df['Gender'] == 'Female') & (df['Tutoring'] == 'NO')]['GPA'].mean(), 2)

print('Average success rate of male students attending private lessons:', m1)
print('Average success rate of male students who do not attend private lessons:', m2,"\n")

print('Average success rate of female students attending private lessons:', f1)
print('Average success rate of female students who do not attend private lessons:', f2,"\n")


genders = ['Male', 'Female']
m = [m1, m2]
f = [f1, f2]

bar_width = 0.23

r1 = np.arange(len(genders))
r2 = [x + bar_width / 2 for x in r1]

plt.bar(r1, m, color = 'darkgray',width = 0.1, label = 'Tutoring')
plt.bar(r2, f, color = 'lightgray',width = 0.1, label = 'No Tutoring')
plt.legend()
plt.tight_layout()
plt.xticks([r + bar_width/2 - 0.06 for r in range(2)], ['Male', 'Female'])
plt.title('GPA VALUES GRAPH BY TUTORING\n')
plt.xlabel('\nGENDER')
plt.ylabel('GPA\n')

plt.show()

# We can understand from the table that the GPA value of those who go to private lessons is higher.
```

```python
averageMaleAbsences0_5 = round(df[(df['Gender'] == 'Male') & ((df['Absences'] >= 0) & (df['Absences'] <= 5))]['GPA'].mean(), 2)
averageMaleAbsences6_10 = round(df[(df['Gender'] == 'Male') & ((df['Absences'] >= 6) & (df['Absences'] <= 10))]['GPA'].mean(), 2)
averageMaleAbsences11_15 = round(df[(df['Gender'] == 'Male') & ((df['Absences'] >= 11) & (df['Absences'] <= 15))]['GPA'].mean(), 2)
averageMaleAbsences16_ = round(df[(df['Gender'] == 'Male') & (df['Absences'] >= 16)]['GPA'].mean(), 2)

averageFemaleAbsences0_5 = round(df[(df['Gender'] == 'Female') & ((df['Absences'] >= 0) & (df['Absences'] <= 5))]['GPA'].mean(), 2)
averageFemaleAbsences6_10 = round(df[(df['Gender'] == 'Female') & ((df['Absences'] >= 6) & (df['Absences'] <= 10))]['GPA'].mean(), 2)
averageFemaleAbsences11_15 = round(df[(df['Gender'] == 'Female') & ((df['Absences'] >= 11) & (df['Absences'] <= 15))]['GPA'].mean(), 2)
averageFemaleAbsences16_ = round(df[(df['Gender'] == 'Female') & (df['Absences'] >= 16)]['GPA'].mean(), 2)

AbsenteeismNumbers = ['[0-5]', '[6-10]', '[11-15]', '[>16]']

MaleAbsenteeismNumbers = [averageMaleAbsences0_5, averageMaleAbsences6_10, averageMaleAbsences11_15, averageMaleAbsences16_]
FemaleAbsenteeismNumbers = [averageFemaleAbsences0_5, averageFemaleAbsences6_10, averageFemaleAbsences11_15, averageFemaleAbsences16_]

plt.bar(AbsenteeismNumbers, MaleAbsenteeismNumbers, width = 0.4)
plt.title("MALE'S ABSENTEEISM-SUCCESSFUL GRAPH")
plt.xlabel("\nNUMBER OF ABSENCES")
plt.ylabel("GPA\n\n")
plt.show()

plt.bar(AbsenteeismNumbers, FemaleAbsenteeismNumbers, width = 0.4, color = 'hotpink')
plt.title("\n\nFEMALE'S ABSENTEEISM-SUCCESSFUL GRAPH")
plt.xlabel("\nNUMBER OF ABSENCES")
plt.ylabel("GPA\n")
plt.show()

# We can observe that as absenteeism increases, the GPA value decreases.
```

```python
sportsNumber = len(df[df['Sports'] == 1])
musicNumber = len(df[df['Music'] == 1])

activities = ['Sports', 'Music']

plt.bar(activities, [sportsNumber, musicNumber], width = 0.2, color = 'green')
plt.title("ACTIVITY GRAPH\n")
plt.xlabel("\nACTIVITIES")
plt.ylabel("NUMBER\n")
plt.show()

maleSportsNumber = len(df[(df['Gender'] == 'Male') & (df['Sports'] == 1)])
maleMusicNumber = len(df[(df['Gender'] == 'Male') & (df['Music'] == 1)])

plt.bar(['Sports','Music'], [maleSportsNumber, maleMusicNumber], width = 0.2)
plt.title("\n\nACTIVITY GRAPH FOR MALE STUDENTS")
plt.xlabel("\nACTIVITY")
plt.ylabel("NUMBER\n")
plt.show()

femaleSportsNumber = len(df[(df['Gender'] == 'Female') & (df['Sports'] == 1)])
femaleMusicNumber = len(df[(df['Gender'] == 'Female') & (df['Music'] == 1)])

plt.bar(['Sports','Music'], [femaleSportsNumber, femaleMusicNumber], width = 0.2)
plt.title("\n\nACTIVITY GRAPH FOR FEMALE STUDENTS")
plt.xlabel("\nACTIVITY")
plt.ylabel("NUMBER\n")
plt.show()
```

```python
Activity_and_GPA = df[((df['Sports'] > 0) | (df['Music'] > 0))]['GPA'].mean()
NoActivity_and_GPA = df[((df['Sports'] == 0) & (df['Music'] == 0))]['GPA'].mean()

print("\nGRADE AVERAGE OF STUDENTS WHO DID THE ACTIVITY:", round(Activity_and_GPA, 2))
print("GRADE AVERAGE OF STUDENTS WHO DO NOT DO ACTIVITIES:", round(NoActivity_and_GPA, 2),"\n")
```
